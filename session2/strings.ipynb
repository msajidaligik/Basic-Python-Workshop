### 2. Accessing Substrings from a String

Two ways:
- Indexing
- Slicing

#### 1. Indexing

```python
# Concept of Indexing
c = "hello"
print(c)
```

```python
print(c[0])
```

```python
print(c[4])
```

Types of Indexing:
- Positive indexing => from left to right, having +ve value inside square brackets
- Negative indexing => from right to left, having -ve value inside square brackets

```python
#Positive indexing
print(c[3])
```

```python
#Negative indexing
print(c[-5])
```

#### 2. Slicing

```python
c = "Hello World"
print(c)
```

```python
print(c[0:5])
```

```python
print(c[2:])
```

```python
print(c[:4])
```

```python
print(c[:])
```

```python
print(c)
```

```python
print(c[0:8:3])
```

```python
print(c[0:6:-1])
```

```python
print(c[-5:-1:2])
```

#### format

```python
"Hello my name is {} and I am {}".format("Ali", 25)
```

```python
"Hello my name is {0} and I am {1}".format("Ali", 25)
```

```python
"Hello my name is {1} and I am {0}".format(25, "Ali")
```

```python
"Hello my name is {name} and I am {age}".format(name = "Ali", age = 25)
```

```python
"Hello my name is {age} and I am {name}".format(name = "Ali", age = 25)
```
